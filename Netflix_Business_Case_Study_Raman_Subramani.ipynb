**How many days after a movie's release will it be added to Netflix?**

```python
# Converting 'date_added' and 'release_year' to datetime for calculation.

nf['date_added'] = pd.to_datetime(nf['date_added'], errors='coerce')
nf['release_date'] = pd.to_datetime(nf['release_year'], format='%Y')

# Calculating the difference in days between 'date_added' and 'release_date'.

nf['days_to_add'] = (nf['date_added'] - nf['release_date']).dt.days
print(nf['days_to_add'])
```

**What is the best time to launch a Movie or a TV show?**

**Best day of the week to release the TV-show or the Movie.**

```python
# Extracting the day of the week from the 'date_added' column.

nf['day_of_week_added'] = nf['date_added'].dt.day_name()

# Counting the number of unique shows added on each day of the week.

shows_per_day_of_week = nf.groupby('day_of_week_added')['show_id'].nunique()

# Finding the most popular day to release a Movie/TV show on Netflix.

most_popular_day = shows_per_day_of_week.idxmax()
most_popular_day_count = shows_per_day_of_week.max()

# Preparing data for the pie chart.

days = shows_per_day_of_week.index
counts = shows_per_day_of_week.values

plt.figure(figsize=(8, 5))
plt.pie(counts, labels = days, autopct=lambda p: f'{p:.1f}%', startangle= 90)
plt.title('Distribution of Netflix Shows Releases by Day of the Week')
plt.legend(loc='upper right', bbox_to_anchor=(1.8, 1))
plt.show()
```

**Most Popular Day for Releases:** Analyzing the pie chart can provide insights that Friday being the week-end sees increase in viewrship. This implies that audiences tends to spend their leisure time by solo/group watching TV Shows or Movies.

**Find which is the best week to release the TV-show or the movie? Do the analysis separately for TV-shows and Movies?**

```python
# Converting 'date_added' to datetime and extracting week number.

nf['date_added'] = pd.to_datetime(nf['date_added'], errors='coerce')
nf['week_number'] = nf['date_added'].dt.isocalendar().week

# Separating Movies and TV Shows.

movies_data = nf[nf['type'] == 'Movie']
tv_shows_data = nf[nf['type'] == 'TV Show']

# Grouping by week number and counting unique titles for Movies & TV Shows.

movies_weekly = movies_data.groupby('week_number')['title'].nunique().reset_index()
tv_shows_weekly = tv_shows_data.groupby('week_number')['title'].nunique().reset_index()

movies_weekly
tv_shows_weekly

# IF we runs the movies_weekly and tv_shows_weekly, we will get a Week number(Means 1st week, 2nd week,....) and aswell we get a unique titles for movies and Tv shows
```

```python
# Identifying the best week for releasing Movies.

best_week_movies = movies_weekly.sort_values('title', ascending=False).iloc[0]
best_week_movies
```

```python
# Identifying the best week for releasing TV shows.

best_week_tv_shows = tv_shows_weekly.sort_values('title', ascending=False).iloc[0]
best_week_tv_shows
```

**Trends in Viewing Habits:** The decision of releasing Movies in 1st week and TV shows in the 27th week of the year influenced by viewing habits of the subscribers. This could be influenced by the factors like get-togethers, holidays, weather, school vacations. For instance, beginning of the year sees new year holidays & get-togethers, winter season which potentially leads to increase in viewership of Movies. The 27th week coincides with school vacations, half-yearly breaks in various regions, increasing viewership of TV Shows as people have more leisure time.

**Find which is the best month to release the TV-show or the movie. Do the analysis separately for TV-shows and Movies?**

```python
# Filtering out TV shows and movies, creating independent copies.

tv_shows_data = nf[nf['type'] == 'TV Show'].copy()
movies_data = nf[nf['type'] == 'Movie'].copy()

# Adding 'month' columns to both dataframes.

# Convert the 'date_added' column to datetime format

movies_data['date_added'] = pd.to_datetime(movies_data['date_added'], errors='coerce')
tv_shows_data['date_added'] = pd.to_datetime(tv_shows_data['date_added'], errors='coerce')

# Now, you extract the month

movies_data['month'] = movies_data['date_added'].dt.month
tv_shows_data['month'] = tv_shows_data['date_added'].dt.month


# Grouping by month and counting unique titles for Movies & TV Shows.

movies_monthly_count = movies_data.groupby('month')['title'].nunique().reset_index()
movies_monthly_count
tv_shows_monthly_count = tv_shows_data.groupby('month')['title'].nunique().reset_index()
tv_shows_monthly_count
```

```python
# Identifying the best month for releasing Movies.

best_month_movies = movies_monthly_count.sort_values('title', ascending=False).iloc[0]
best_month_movies
```

```python
# Identifying the best month for releasing TV Shows.

best_month_tv_shows = tv_shows_monthly_count.sort_values('title', ascending=False).iloc[0]
best_month_tv_shows.astype('int')
```

**Targeted Monthly Releases:** The decision of releasing Movie and TV show in July to December conincides with the holiday season when audiences are more available to watch new contents. **July beings the summer holiday month, half-yearly breaks in many regions and December being the new year and winter break period(X-mas Holidays)**, is likely the time when family and group viewership increases, making it an ideal time for releasing Movies & TV shows which can persuade audiences for binge-watching.

**Counting unique titles per year for each category.**

```python
# Filtering Total data for Movies, TV Shows.

# Total Data for Movies
movies_data = nf[nf['type'] == 'Movie'].copy()
movies_data
```

```python
# Total Data for TV shows

tv_shows_data = nf[nf['type'] == 'TV Show'].copy()
tv_shows_data
```

```python
# Convert 'date_added' to datetime and extract the year for all DataFrames

movies_data['year_added'] = pd.to_datetime(movies_data['date_added'], errors='coerce').dt.year
tv_shows_data['year_added'] = pd.to_datetime(tv_shows_data['date_added'], errors='coerce').dt.year
nf['year_added'] = pd.to_datetime(nf['date_added'], errors='coerce').dt.year

# Counting unique titles per year for each category

unique_movies_per_year = movies_data.groupby('year_added')['show_id'].nunique()
unique_tv_shows_per_year = tv_shows_data.groupby('year_added')['show_id'].nunique()
unique_total_per_year = nf.groupby('year_added')['show_id'].nunique()

# Ensure the indices are consistent across all data

all_years = unique_total_per_year.index.union(unique_movies_per_year.index).union(unique_tv_shows_per_year.index)

unique_movies_per_year = unique_movies_per_year.reindex(all_years, fill_value=0)
unique_tv_shows_per_year = unique_tv_shows_per_year.reindex(all_years, fill_value=0)
unique_total_per_year = unique_total_per_year.reindex(all_years, fill_value=0)

# Plotting the data

plt.figure(figsize=(10, 5))
plt.plot(unique_movies_per_year, label='Movies')
plt.plot(unique_tv_shows_per_year, label='TV Shows')
plt.plot(unique_total_per_year, label='Total')
plt.title('Number of Unique Titles Added to Netflix Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Shows')
plt.legend()
plt.grid(True)
plt.show()
```

**Un-nesting the column**

```python
# Function for handling un-nesting.

def unnest_dataframe(df, column):
    return (
        df.drop(column, axis=1)
        .join(
            df[column].str.split(',', expand=True)
            .stack()
            .reset_index(level=1, drop=True)
            .rename(column)
        )
    )
```

```python
# Un-nesting the Columns

# Un-nesting the director's column:
unnested_director = unnest_dataframe(nf, 'director')

#Un-nesting the cast's column:
unnested_cast = unnest_dataframe(nf, 'cast')

#un-nesting the county's column:
unnested_countries = unnest_dataframe(nf, 'country')

#un-nesting the listed_in column:
unnested_listed_in = unnest_dataframe(nf, 'listed_in')
```

**Creating a histogram for the duration of movies**

```python
# Extracting the numeric part of the duration for movies.

# Create a copy to avoid the warning.

movies_data = nf[nf['type'] == 'Movie'].copy()
movies_data['duration_numeric'] = movies_data['duration'].str.extract('(\d+)')

# Convert to float first to handle NaN values.

movies_data['duration_numeric'] = movies_data['duration_numeric'].astype(float)

# Dropping NaN values.

movies_data = movies_data.dropna(subset=['duration_numeric'])

# Now safely convert to integer.

movies_data['duration_numeric'] = movies_data['duration_numeric'].astype(int)

# Creating a histogram for the duration of movies.

plt.figure(figsize=(8, 6))
sns.histplot(movies_data['duration_numeric'], bins=30)
plt.title('Histogram of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

**Total value counts for the number of seasons in the 'duration' column for TV shows.**

```python
# Performing value counts on the number of seasons in the 'duration' column for TV shows and ensuring here I'm only considering unique TV shows.

# Filtering out TV shows and creating an independent copy.

tv_shows_data = nf[nf['type'] == 'TV Show'].copy()

# Extracting the number of seasons.

tv_shows_data['seasons_count'] = tv_shows_data['duration'].str.extract('(\d+)').astype(float)

# Dropping duplicates to ensure each TV show is counted only once.

unique_tv_shows = tv_shows_data.drop_duplicates(subset='show_id')

# Value counts of the number of seasons.

seasons_counts = unique_tv_shows['seasons_count'].value_counts().sort_index()

# Creating a bar plot for the value counts of the number of seasons in TV shows.

plt.figure(figsize=(10, 6))
seasons_counts.plot(kind='bar', color='purple')
plt.title('Distribution of the Number of Seasons for TV Shows on Netflix')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of TV Shows')
plt.xticks(rotation = 45)
plt.show()
```
